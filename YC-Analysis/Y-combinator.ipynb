```python
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
import pylab
plt.style.use('fivethirtyeight')
%matplotlib inline
```

```python
df = pd.read_csv("companies.csv")
```

```python
df.dtypes
```

```python
# Create a new categorical variable with the batch category, either 's' or 'w'
df.batch_type  = df.batch.apply(lambda x: x[:1]).astype('category')
```

```python
df.head()
```

# How many companies started per year
- It looks like an upward trend
- A dip happened in the year 2013

```python
plot = df.name.groupby(df.year).count().plot.line()
```

# Companies per batch type
- Generally more companies in the summer batch than in the winter batch

```python
plot = df.name.groupby(df.batch_type).count().plot.bar()
```

# Companies per year per batch type
- 2005 had only a summer batch
- From 2009 to 2015, the summer batches had more companies

```python
df.name.groupby([df.year, df.batch_type]).count().unstack().plot.bar()
```

# Types of Companies
- B2B and consumer are the most popular
- Education and BioMedical are lower in popularity

```python
df.name.groupby(df.vertical).count().sort_values().plot.bar()
```

# Types of companies per batch type
- Comparable except for Education and Fintech which seem a little more popular in summer

```python
df.name.groupby([df.vertical, df.batch_type]).count().unstack().plot.bar()
```

# Types of Companies over the years
- B2B, consumer and hardware verticals have skyrocketed over the past few

```python
pylab.rcParams['figure.figsize'] = (13, 6)
df.name.groupby([df.year, df.vertical]).count().unstack().plot.line()
```
